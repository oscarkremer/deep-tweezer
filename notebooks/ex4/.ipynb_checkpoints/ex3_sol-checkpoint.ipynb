### Exercise 2.4. 

Reduce the rendering (maybe do it only once every $10$ episodes), and run the algorithm, printing the total reward per episode. It should converge to values above $-300$ in around $50$ episodes. Plot the value function using `env.plot(dqn)`. Is it what you would expect based on the rewards given by the environment? Why (not)?

#### Answer

Implementation in the cells bellow. The plot converged to a similar figure encountered in the last task, however when examining the values of the function in the state space the function showed positive values, which would be impossible considering the reward function only returnes 0 or negative values.

```python
def random_action(s, dqn, epsilon, actions, probs):
    if np.random.random() < epsilon:
        return np.random.choice(actions, p=probs)
    else:
        return actions[dqn(s, actions).argmax()]
```

```python
def deep_q(env, 
           epsilon=1, 
           gamma=0.99,
           episodes=200, 
           steps=1000, 
           batch_size=256, 
           render_step=10, 
           memory_min=1000,     
           rate = 0.97,
           epsilon_min = 0.05,
           hiddens=[25, 25], 
           eps_type='function', 
           plot=True):
    s = env.reset()
    actions = env.actions
    number_actions =actions.shape[0]
    probs = number_actions*[1/number_actions]
    dqn = DQN(s.shape[0], hiddens=hiddens)
    memory = Memory(s.shape[0], 1)
    rewards = []
    for e in tqdm(range(episodes)):
        s = env.reset()
        a = random_action(s, dqn, epsilon, actions, probs)
        total_r = 0
        for t in range(steps):
            sP, r, done, _ = env.step(a)
            total_r+=r
            memory.add(s, a, r, sP, done)
            aP = random_action(sP, dqn, epsilon, actions, probs)
            if memory.__len__() >= memory_min:
                s_m, a_m, r_m, sP_m, done_m = memory.sample(batch_size)
                multi_done = 1-done_m
                target = r_m + gamma*multi_done*(np.max(dqn(sP_m, actions), axis=1).reshape((batch_size,1)))
                dqn.train(s_m, a_m, target)
            if render_step:
                if not t%render_step:
                    env.render()
            s, a = sP, aP    
            if done:
                break
        rewards.append(total_r) 
        if eps_type=='function':
            epsilon = rate*epsilon if epsilon*rate > epsilon_min else epsilon
    env.close()
    if plot:
        env.plot(dqn)
    return dqn, rewards
```

```python
env = Pendulum()
dqn, rewards = deep_q(env)
```

```python
dqn_const, rewards_const = deep_q(env, eps_type='constant', epsilon=0.05, render_step=0)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,4))
plt.plot(rewards, linewidth=2.5)
plt.plot(rewards_const, linewidth=2.5)
plt.legend(['Exploration Schedule', 'Epsilon=0.05'], fontsize=8)   
plt.title("Reward behavior with dqn")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.show()
```

### Exercise 3.2.

Implement a target network by creating a second identical DQN object and copying the original networks weights to it every $50$ timesteps. Use `target <<= dqn` to copy the weights. Use the target network to calculate the target values in Eq 1.The learning should converge to values above $100$ in around $200$ episodes.

#### Answer:

Implementation in cell bellow.

```python
def deep_q_target(env, 
                  epsilon=1, 
                  gamma=0.99,
                  episodes=200, 
                  steps=1000, 
                  batch_size=256, 
                  render_step=10, 
                  memory_min=1000, 
                  rate = 0.97, 
                  epsilon_min = 0.05, 
                  hiddens=[25, 25], 
                  eps_type='function', 
                  target_step=50,
                  plot=True):
    s = env.reset()
    actions = env.actions
    number_actions =actions.shape[0]
    probs = number_actions*[1/number_actions]
    dqn = DQN(s.shape[0], hiddens=hiddens)
    target_network = DQN(s.shape[0], hiddens=hiddens)
    memory = Memory(s.shape[0], 1)
    rewards = []
    for e in tqdm(range(episodes)):
        s = env.reset()
        a = random_action(s, dqn, epsilon, actions, probs)
        total_r = 0
        for t in range(steps):
            sP, r, done, _ = env.step(a)
            total_r+=r
            memory.add(s, a, r, sP, done)
            aP = random_action(sP, dqn, epsilon, actions, probs)
            if memory.__len__() >= memory_min:
                s_m, a_m, r_m, sP_m, done_m = memory.sample(batch_size)
                multi_done = 1-done_m
                target = r_m + gamma*multi_done*(np.max(target_network(sP_m, actions), axis=1).reshape((batch_size,1)))
                dqn.train(s_m, a_m, target)
            if not t%target_step:
                target_network <<= dqn
            if render_step:
                if not t%render_step:
                    env.render()
            s, a = sP, aP    
            if done:
                break
        rewards.append(total_r) 
        if eps_type=='function':
            epsilon = rate*epsilon if epsilon*rate > epsilon_min else epsilon
    env.close()
    if plot:
        env.plot(dqn)
    return dqn, rewards
```

```python
env = Lander()
dqn, r_target_lander = deep_q_target(env, plot=False, render_step=0)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,4))
plt.plot(r_target_lander, linewidth=2.5)
plt.title("Reward behavior during simulation with target network")
plt.xlabel("Time-step")
plt.ylabel("Reward")
plt.show()
```

### Exercise 3.3.

Try to find hyperparameters (discount rate, exploration schedule, batch size, target network update interval, hidden layer parameters) that give more consistent performance. Document your search. Note that the results will be stochastic!

#### Answer:

To run the optimization the following hyperparameters were choosen: `hiddens`, `gamma` and `batch_size`. For the `hiddens` the tested values were `[15, 15]`,`[25, 25]` and `[35, 35]]`, for `gamma` `0.9`, `0.99` and `0.999`, and for the `batch_size` `64`, `128`, `256` and `512`. The search was done in this order, using the best parameter for the next parameter, so first the method find the best `hiddens` parameter, then using the best `hiddens` search for the best `gamma`, and ,lastly, using the best `gamma` and `hiddens` find the optimal `batch_size`. Each combination was tested `3` times, were was used the mean of the total reward in the last 10 episodes. The metric is averaged in all the tests, and the best one is choosed. The number of episodes was reduced to 100 

```python
def aux_optimize(env, params, name, other_kwargs={}, trials=3, mean_window=10):
    metrics = np.zeros((trials, len(params)))
    for j, param in enumerate(params):
        arguments = {'plot': False, name: param, 'render_step': 0, 'episodes': 100}
        arguments.update(other_kwargs)
        for i in range(trials):
            dqn_t_lander, r_target_lander = deep_q_target(env, **arguments)
            metrics[i][j] = np.array(r_target_lander[-mean_window:]).mean()
    return params[metrics.mean(axis=0).argmax()], metrics.mean(axis=0)
```

```python
def optimize():
    env = Lander()
    hiddens = [[15, 15],[25, 25],[35, 35]]
    gamma = [0.9, 0.99, 0.999]
    batch_sizes = [64, 128, 256, 512]
    best_hidden, metric_hidden = aux_optimize(env, hiddens, 'hiddens')
    other_kwargs = {'hiddens': best_hidden}
    best_gamma, metric_gamma = aux_optimize(env, hiddens, 'gamma', other_kwargs=other_kwargs)
    other_kwargs['gamma'] = gamma
    best_batch, metric_batch = aux_optimize(env, batch_sizes, 'batch_size', other_kwargs=other_kwargs)
    other_kwargs['batch_size'] = best_batch
    other_kwargs['metrics'] = {'hiddens': metric_hidden, 'gamma': metric_gamma, 'batch_size': metric_batch}
    return other_kwargs
```

```python
metrics = optimize()
```
