```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error


from ex5 import Pendulum, rbfprojector, Model, Memory

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

def rmse(x,y):
    return mean_squared_error(x,y,squared=False)

def plot_rewards(experiment_rewards):
    fig,ax = plt.subplots(figsize=(8,6))
    plt.plot(range(len(experiment_rewards)),experiment_rewards)
    plt.grid(True)
    plt.xlabel("Episodes")
    plt.ylabel("Cumulative Reward")
    plt.title("Rewards over episodes")
    plt.show()

def plot_errors(df_errors:pd.DataFrame):
    fig,ax = plt.subplots(figsize=(8,6))
    df_errors.set_index('episode').plot(kind='line',ax=ax)
    plt.grid(True)
    plt.ylabel("RMSE")
    plt.xlabel("Episodes")
    plt.title("Rewards and States Erros")
    plt.show()

help(Model)
```

```python
from abc import ABC, abstractclassmethod
from time import time

class ModelApproximationExperiment(ABC):
    
    '''
    Abstract class that implements a generic model approximation experiment. 

    Experiments should be run on the run_experiment or the evaluate_experiments methods.

    These functions are already defined, only the intermediate building blocks should be implemented.

    These blocks are defined by the abstract methods.
    '''

    def __init__(self,
                env,
                error_function=rmse,
                sigma = 1,
                gamma=0.99,
                exploration_schedule = 0.99): 
        self.env = env
        self.sigma = sigma
        self.gamma = gamma
        self.exploration_schedule = exploration_schedule
        self.error_function = error_function
        
    def run_experiment(self,max_episodes=100,simulated_episodes=10,show_render = False,enable_learning=True,use_transitions=True) -> tuple:
        '''
        Runs the RL experiment. This function can be used for training or testing, depending on the enable_learning flag.

        Parameters
        ----------

        max_episodes: The number of episodes to run the experiment.

        show_render: Flag to show (or not) the pendulum environment.

        enable_learning: Flag to enable (or not) the learning of the agent.

        use_transitions: Flag to enable model on transitions between states rather than states themselves.

        Returns
        --------

        experiment_rewards: Array containing the cumulative rewards of every episode,
        
        experiment_estimate_errors: Pandas Dataframe containing the estimate errors for rewards and next states,

        computation_time: The time that the experiment took to run, in seconds,

        rise_time: How many episodes did it take for the agent to obtain cumulative rewards above -500. 

        '''

        self.memory = Memory(states=self.env.states,actions=1)

        if enable_learning:
            self.model = Model(states=self.env.states,actions=1,diff=use_transitions)
            self._initialize_experiment_variables()
            print("Experiment initialization...")
    
        start_time = time()
        sigma = self.sigma
        experiment_rewards = []
        simulated_experiment_rewards = []
        experiment_estimate_errors = []
        rise_time = 1

        for episode in range(max_episodes):

            real_episode_returns = self._real_episode_iterations(enable_learning,show_render,episode)
            cumulative_reward,next_states_tracking,rewards_tracking,next_estimate_states_tracking,rewards_estimate_tracking = real_episode_returns

            if enable_learning:
                if episode%10 == 0:
                    print(f"EPISODE {episode + 1} - CUMULATIVE REWARD: {cumulative_reward}") 
                    print("Updating agent...")
                

            if simulated_episodes > 0:
                experiment_estimate_errors.append(self._calculate_model_errors(
                    next_estimate_states_tracking,next_states_tracking,
                    rewards_estimate_tracking,rewards_tracking,
                    episode))
                
                if episode == 0:
                    print("Starting model simulation training...\n")

                if episode%10 == 0 or episode < 20:#reducing the amount of model training after 15 episodes.
                    self._train_environment_model()
                
                if episode >=20:
                    print("Starting model simulation episodes...\n")
                    simulated_cumulative_rewards,sigma = self._simulate_episodes_iterations(sigma=sigma,enable_learning=enable_learning,simulated_episodes=simulated_episodes)
                    simulated_experiment_rewards.append(simulated_cumulative_rewards)

            experiment_rewards.append(cumulative_reward)
            sigma = sigma * self.exploration_schedule
            
            if cumulative_reward < -500: # Registering the last episode that the cumulative is less than -500.
                rise_time = episode+1 

        self.env.close()
        computation_time = time() - start_time

        # TODO: Improve this return, put everythin in a DataFrame
        experiment_estimate_errors = pd.DataFrame(experiment_estimate_errors)
        return experiment_rewards,simulated_experiment_rewards,experiment_estimate_errors,computation_time,rise_time

    def evaluate_experiments(self,n_times=10,simulated_episodes=10,max_episodes=100) -> tuple:
        
        rise_time_evaluation = []
        end_performance_train = []
        simulated_end_performance_train = []
        end_performance_evaluation = []
        computation_time_evaluation = []

        for ii in range(n_times):
            print(f"\nRunning experiment {ii+1}/{n_times}")
            print("Learning...")
            cumulative_rewards_train,simulated_cumulative_rewards_train,_,computation_time,rise_time = self.run_experiment(max_episodes=max_episodes, enable_learning=True,show_render=False)

            print("Testing...")
            cumulative_rewards,_,_,_,_ = self.run_experiment(max_episodes=max_episodes,simulated_episodes=simulated_episodes, enable_learning=False,show_render=False)

            print(f"Rise time (episodes):                {rise_time} \n"+
                   "Computation time (seconds):          %.3f  \n" % computation_time+
                   "Mean cumulative reward (train):      %.3f  \n" % np.mean(cumulative_rewards_train)+
                   "Mean cumulative reward (train simulation):      %.3f  \n" % np.mean(simulated_cumulative_rewards_train)+
                   "Mean cumulative reward (test):       %.3f" % np.mean(cumulative_rewards))

            rise_time_evaluation.append(rise_time)
            computation_time_evaluation.append(computation_time)
            end_performance_evaluation.append(np.mean(cumulative_rewards))
            end_performance_train.append(np.mean(cumulative_rewards_train))
            simulated_end_performance_train.append(np.mean(simulated_cumulative_rewards_train))

        print("\nFinished all experiments! \nPerformance metrics in mean(std).\n" +
                "Rise time (episodes):       %.3f (%.3f)\n" % (np.mean(rise_time_evaluation),np.std(rise_time_evaluation))+
                "Computation time (seconds): %.3f (%.3f)\n" % (np.mean(computation_time_evaluation),np.std(computation_time_evaluation))+
                "Mean cumulative reward :    %.3f (%.3f)\n" % (np.mean(end_performance_evaluation),np.std(end_performance_evaluation)))
        
        return rise_time_evaluation, computation_time_evaluation, end_performance_evaluation,end_performance_train,simulated_end_performance_train

    def _real_episode_iterations(self,enable_learning,show_render,episode):

        current_state = self.env.reset()
        cumulative_reward = 0
        next_states_tracking = []
        rewards_tracking = []
        next_estimate_states_tracking = []
        rewards_estimate_tracking = []

        for t in range(1000):
            
            action = self._choose_action(current_state, sigma, enable_learning)

            next_state,reward,done,_ = self.env.step(action)
            estimate_reward,next_estimate_state = self.model(current_state,action)

            next_estimate_states_tracking.append(next_estimate_state)
            rewards_estimate_tracking.append(estimate_reward)

            next_states_tracking.append(next_state)
            rewards_tracking.append(reward)
            
            self.memory.add(s=current_state,a=action,r=reward,sp=next_state,done=done)
            self.time_step = t
            
            if show_render and episode%10 == 0:
                self.env.render()

            if done:
                break
            
            if enable_learning:
                self._update_agent(current_state, next_state, reward, action)

            current_state = next_state
            cumulative_reward += reward
        
        return cumulative_reward,next_states_tracking,rewards_tracking,next_estimate_states_tracking,rewards_estimate_tracking

    def _train_environment_model(self) -> None:
        self.model.fit(memory=self.memory)

    def _calculate_model_errors(self,next_estimate_states_tracking,next_states_tracking,rewards_estimate_tracking,rewards_tracking,episode) -> dict:
        episode_state_error = self.error_function(next_estimate_states_tracking,next_states_tracking)
        episode_reward_error = self.error_function(rewards_estimate_tracking,rewards_tracking)
        return {'episode':episode + 1,'states_error':episode_state_error,'rewards_error':episode_reward_error}
        
    def _simulate_episodes_iterations(self,sigma,simulated_episodes,steps=20, enable_learning=True,show_render=False):
        
        simulated_cumulative_reward = 0
        for episode in range(simulated_episodes):
            current_state = self.env.reset()
            # sigma = self.exploration_schedule
            for step in range(steps):

                action = self._choose_action(current_state, sigma, enable_learning = enable_learning)
                reward, next_state = self.model(s=current_state,a=action)
                next_state = env.normalize(next_state)
                
                if self._state_is_out_of_bounds(next_state):
                    print("Breaking simulated episode! Model simulation output is out of bounds.")
                    break
                
                if enable_learning:
                    self._update_agent(current_state, next_state, reward, action)

                self.memory.add(s=current_state,a=action,r=reward,sp=next_state,done=False) #how do I manually know when the episode is done? Check if it is in the final position?
                current_state = next_state
                sigma = sigma * self.exploration_schedule
            
                if show_render and episode%10 == 0:
                    self.env.render()

                current_state = next_state
                simulated_cumulative_reward += reward
            
        return simulated_cumulative_reward,sigma

    def _state_is_out_of_bounds(self,state):
        return not np.all(state > env.env.observation_space.low) and np.all(state < env.env.observation_space.high)
    
    @abstractclassmethod
    def _initialize_experiment_variables(self) -> None:
        pass

    @abstractclassmethod
    def _choose_action(self,current_state:list, sigma:float, enable_learning:bool) -> float:
        pass

    @abstractclassmethod
    def _update_agent(self,current_state:list, next_state:list, reward:float, action:float) -> None:
        pass

class LinearStateValueActorCritic(ModelApproximationExperiment):

    '''
    This class is the implementation of the Linear State Value Actor Critic model.
    '''

    def __init__(self,env,feature,
                sigma=1,
                gamma=0.99,
                alpha=0.2,
                beta=0.01,
                exploration_schedule=0.99):
        
        super().__init__(env=env,
                        sigma=sigma,
                        gamma=gamma,
                        exploration_schedule=exploration_schedule,
                        error_function=rmse)
        self.feature = feature
        self.alpha = alpha
        self.beta = beta

    def _initialize_experiment_variables(self):
        self.w = np.ones(len(self.feature(np.ones(self.env.states)))) * 0.1
        self.theta = np.ones(len(self.feature(np.ones(self.env.states)))) * 0.1

    def _choose_action(self,current_state,sigma, enable_learning):

        pi_mean = np.dot(self.feature(current_state).T,self.theta) # phi(s)ˆT * theta
        
        if enable_learning:
            
            action = np.random.normal(loc=pi_mean,scale=sigma) # sampling from a gaussian distribution with mean pi_mean and std sigma.
        else:
            action = pi_mean # deterministic action after training.
        return action

    def _update_agent(self,current_state, next_state, reward, action):
        delta = reward + (self.gamma * np.dot(self.feature(next_state).T, self.w)) - np.dot(self.feature(current_state).T,self.w)

        self.w = self.w + self.alpha * self.feature(current_state) * delta

        self.theta = self.theta + self.beta * (action - np.dot(self.feature(current_state).T,self.theta)) * self.feature(current_state) * delta
```

```python
gamma = 0.99
sigma = 1
exploration_schedule = 0.99
alpha = 0.2
beta = 0.01

n_functions = 21
sigma_rbf = 0.25
feature = rbfprojector(nbasis=n_functions,sigma=sigma_rbf)

env = Pendulum()

actor_critic_experient = LinearStateValueActorCritic(env,feature,
                                                    exploration_schedule = exploration_schedule,
                                                    sigma=sigma,
                                                    gamma=gamma,
                                                    alpha=alpha,
                                                    beta=beta)

experiments_results = actor_critic_experient.run_experiment(max_episodes = 100,simulated_episodes=50, show_render = True, use_transitions = False)
experiment_rewards,simulated_experiment_rewards,experiment_estimate_errors,computation_time,rise_time = experiments_results
```
